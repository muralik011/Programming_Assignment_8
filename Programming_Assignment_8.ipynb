1. Write a Python Program to Add Two Matrices?

```python
import numpy as np
```

```python
m1 = np.array([[1, 2, 3], [4, 5, 6]], dtype='int8')
m2 = np.array([[11, 22, 33], [44, 55, 66]], dtype='int8')
print(f'm1\n{m1}\n')
print(f'm2\n{m2}\n')
print(f'm1 + m2\n{m1 + m2}')
```

2. Write a Python Program to Multiply Two Matrices?

```python
m1 = np.array([[1, 2, 3], [4, 5, 6]], dtype='int8')
m2 = np.array([[11, 22], [33, 44], [55, 66]], dtype='int8')
print(f'm1\n{m1}\n')
print(f'm2\n{m2}\n')
print(f'm1 x m2\n{np.dot(m1, m2)}')
```

3. Write a Python Program to Transpose a Matrix?

```python
m1 = np.array([[11, 22], [33, 44], [55, 66]], dtype='int8')
print(f'm1\n{m1}\n')
print(f'Transpose of m1\n{m1.T}')
```

4. Write a Python Program to Sort Words in Alphabetic Order?

```python
l = ['this', 'is', 'array', 'sort']
l.sort()
print(l)
```

5. Write a Python Program to Remove Punctuation From a String?

```python
import re
s = 'hello world! remove punctuations like ,.()[]{}^%&*'
print(s)
s = re.sub("[^\w\s]", "", s)
print(s)
```
